R^2 is not the relable metric to take decision wheather to select feature or not.

```python
import statsmodels.formula.api as smf
lm = smf.ols(formula= 'Sales ~ TV',data=df).fit()
lm.summary()
```

```python
lm = smf.ols(formula= 'Sales ~ TV+Radio',data=df).fit()
lm.summary()
```

```python
lm = smf.ols(formula= 'Sales ~ TV+Radio+Newspaper',data=df).fit()
lm.summary()
```

**null Hypothesis :** feature has no significance.

**Alternate Hypothesis:** Feature has significance.

p-value tells you about the fature is significant or not.

If we observe the p-value for newspaper feature which is greater than 0.05, we will accept the null hypothesis, which is not significant feature,We can remove the newspaper feature.

**Ajd R2** : If the features are increasing that doesn't mean that model is getting better.
    
 In R2 we haven't included any features information so it is not good metric about taking descisions on feature importance.
